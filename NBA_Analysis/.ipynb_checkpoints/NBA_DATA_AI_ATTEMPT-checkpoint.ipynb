```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Player1\Desktop\Creations\dataScience\NBA_Python_Data_Project\Csv_Data\Lebron_Data.csv")
df1 = pd.read_csv(r"C:\Users\Player1\Desktop\Creations\dataScience\NBA_Python_Data_Project\Csv_Data\Team_Defense_Rating.csv")
```

```python
df['HOME=0,AWAY=1'] = df['MATCHUP'].apply(lambda x: 1 if '@' in x else 0)
df['MATCHUP'] = df['MATCHUP'].str.replace('@','', regex = False)
df = pd.merge(df, df1, on='MATCHUP', how='inner')
df = df[['HOME=0,AWAY=1','SPREAD','PROP_LINE','DEF_RATING','PTS']]
df = df[~df['PROP_LINE'].str.contains('NL')]
print(df)
```

```python
def modify_spread(spread):
    number = float(spread.split()[1])
    if 'LAL' in spread:
        return number
    else:
        return -number
df['SPREAD'] = df['SPREAD'].apply(modify_spread)

df['PTS'] = df['PTS'].apply(lambda x: 0 if 'O' in x else 1)

print(df)
```

```python
# y attribute to predict is Over or under Points
# x attributes are all aspects

x = df.drop(columns = ['PTS'])
```

```python
y = df['PTS']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5) 
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential()
print(x_train.shape)
```

```python
model.add(tf.keras.layers.Input(shape =(4,)))
model.add(tf.keras.layers.Dense(256, activation = 'sigmoid'))
model.add(tf.keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=500)
```

```python
model.evaluate(x_test, y_test)
```
